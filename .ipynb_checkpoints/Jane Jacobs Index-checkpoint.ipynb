```python
def get_county_tracts(state_abbr, county_code):
    state_shapefile=gpd.read_file(states.lookup('{}').shapefile_urls('tract')).format(state_abbr)
    county_shapefile=state_shapefile.loc[state_shapefile['COUNTYFP10']==county_code]
    return county_shapefile
```

```python
population=pd.DataFrame(c.acs5.state_county_tract(('NAME', 'B01003_001E'), states.ID.fips,'001',Census.ALL))
population.rename(columns={'B01003_001E':'TotalPopulation'},inplace=True)
population.head()
```

```python
housing_40_49=pd.DataFrame(c.acs5.state_county_tract('B25034_010E', states.ID.fips,'001',Census.ALL))
housing_40_49.rename(columns={'B25034_010E':'1940-1949'},inplace=True)
housing_pre_39=pd.DataFrame(c.acs5.state_county_tract('B25034_011E', states.ID.fips,'001',Census.ALL))
housing_pre_39.rename(columns={'B25034_011E':'Before 1939'},inplace=True)
housing_50_59=pd.DataFrame(c.acs5.state_county_tract('B25034_009E', states.ID.fips,'001',Census.ALL))
housing_50_59.rename(columns={'B25034_009E':'1950-1959'},inplace=True)
housing_60_69=pd.DataFrame(c.acs5.state_county_tract('B25034_008E', states.ID.fips, '001', Census.ALL))
housing_60_69.rename(columns={'B25034_008E':'1960-1969'},inplace=True)
housing_70_79=pd.DataFrame(c.acs5.state_county_tract('B25034_007E', states.ID.fips, '001', Census.ALL))
housing_70_79.rename(columns={'B25034_007E':'1970-1979'},inplace=True)
housing_80_89=pd.DataFrame(c.acs5.state_county_tract('B25034_006E', states.ID.fips, '001', Census.ALL))
housing_80_89.rename(columns={'B25034_006E':'1980-1989'}, inplace=True)
housing_90_99=pd.DataFrame(c.acs5.state_county_tract('B25034_005E', states.ID.fips, '001', Census.ALL))
housing_90_99.rename(columns={'B25034_005E':'1990-1999'},inplace=True)
housing_00_10=pd.DataFrame(c.acs5.state_county_tract('B25034_004E', states.ID.fips, '001', Census.ALL))
housing_00_10.rename(columns={'B25034_004E':'2000-2010'}, inplace=True)
```

```python
housing=housing_pre_39.merge(housing_40_49).merge(housing_50_59).merge(housing_60_69).merge(housing_70_79).merge(housing_80_89).merge(housing_90_99).merge(housing_00_10)
housing=housing[['tract','Before 1939','1940-1949','1950-1959','1960-1969','1970-1979','1980-1989','1990-1999','2000-2010']]
```

```python
housing['std']=housing.std(axis=1)
```

```python
housing.reset_index(inplace=True)
ada_county_housing=pd.merge(ada_county,housing,left_on='TRACTCE10', right_on='tract')
ada_county_housing.plot(column='std',legend=True)
```
